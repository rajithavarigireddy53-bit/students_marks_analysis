# students marks analysis using numpy




#import numy as np


```python
import numpy as np
```

# creating student marks array

```python
marks=np.array([78,98,65,89,95,75,85])
print("student marks:")
print(marks)
```

# find total marks of a student

```python
total=np.sum(marks)
print("total marks:",total)
```

# finding average marks of a student


```python
average=np.mean(marks)
print("average marks:",average)
```

finding highest and lowest marks of a student

```python
highest=np.max(marks)
lowest=np.min(marks)
print("highest marks",highest)
print("lowest marks",lowest)


```

# sort marks


```python
sorted_marks=np.sort(marks)
print("sorted marks:")
print (sorted_marks)
```

# find passed students

```python
passed_students=marks[marks>=70]
print("passed students marks':")
print(passed_students)
```
